### Image Class

```python
#Image Class
#It contains all the methods required


class Image:
    def __init__(self,image = None):                              
        self.name = name
        self.image = image
        print(f"Hello {self.name}\nPlease Choose What You want to do .....")
        self.choiceMenu()
        
        
    def choiceMenu(self):                                            #Main Menu
        print("""1   >>>   Load Image
2   >>>   Edit Image
3   >>>   Save Image
4   >>>   Exit""")
        choice = int(input("\nPlease Enter your Choice : "))
        if choice == 1:
            self.loadImage()
        elif choice == 2:
            self.EditImage()
        elif choice == 3:
            self.SaveImage()
        elif choice == 4:
            print(f"Thank You !!\nHave a Nice Day {self.name}\n")
            exit()
        else:
            print("Please enter valid choice ....")
            self.choiceMenu()
            
            
    def loadImage(self):
        self.image = io.imread('nfs.jpg')                     # inputed image path should be given correctly
        plt.imshow(self.image)
        plt.show()
        print(type(self.image))
        print(self.image.shape)                               # print the shape of the image
        self.choiceMenu()
        
        
    def EditImage(self):                                      #Method for Editing Image , user has to input choice
        print("""PLease Choose what you want to do : 
1   >>>   Vertical Flip
2   >>>   Horizontal Flip
3   >>>   Invert Colors
4   >>>   Gray Scale
5   >>>   Return To Main Menu""")
        Choice1 = int(input("Your Choice : "))
        if Choice1 == 1:
            print("\nFlipping the Image VERTICALLY .... \n")
            self.VerticalFlip()
        elif Choice1 == 2:
            print("\nFlipping the Image HORIZONTALLY .... \n")
            self.HorizontalFlip()
        elif Choice1 == 3:
            print("\nInverting Colors of the image .... \n")
            self.InvertColors()
        elif Choice1 == 4:
            print("\nConverting image to Grayscale .... \n")
            self.GrayScale()
        elif Choice1 == 5:
            print("\nRedirecting to Main Menu\n")
            self.choiceMenu()
        else:
            print("\nPlease Enter valid choice ....\n")
            self.EditImage()
            
            
    def VerticalFlip(self):                                    #Method for Vertical flip of Image
        #vertical = np.flipud(self.image)
        vertical = self.image[::-1,:]
        plt.imshow(vertical)
        plt.show()
        choice2 = input("""\nSave ?                            
Yes  /   No
Please Enter your Choice : """)
        if choice2 == 'yes' or choice2 == 'Yes':                #Saving after every edit to minimize loss
            self.image = vertical
        self.EditImage()
        
        
    def HorizontalFlip(self):                                    #Method for Horizontal Flip
        #horizontal = np.fliplr(self.image)
        horizontal = self.image[:,::-1]
        plt.imshow(horizontal)
        plt.show()
        choice2 = input("""\nSave ?
Yes  /   No
Please Enter your Choice : """)
        if choice2 == 'yes' or choice2 == 'Yes':
            self.image = horizontal
        self.EditImage()
        
        
    def InvertColors(self):                                       #Method for Inverting Colors
        #invert = np.invert(self.np_image)
        invert_colors = np.invert((self.image)[:,:,:3])
        plt.imshow(invert_colors)
        plt.show()
        choice2 = input("""\nSave ?
Yes  /   No
Please Enter your Choice : """)
        if choice2 == 'yes' or choice2 == 'Yes':
            self.image = invert_colors
        self.EditImage()
        
        
    def GrayScale(self):                                           #Method for Grayscale conversion
        def rgb2gray(rgb):
            return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
        GrayScale_img = rgb2gray(self.image)
        plt.imshow(GrayScale_img , cmap=plt.get_cmap('gray'))
        plt.show()
        choice2 = input("""\nSave ?
Yes  /   No
Please Enter your Choice : """)
        if choice2 == 'yes' or choice2 == 'Yes':
            self.image = GrayScale_img
        self.EditImage()
    
    
    def SaveImage(self):
        uint8image = np.uint8(self.image)
        io.imsave(fname='C:\\Users\\Achyut\\bvb1.png',arr = uint8image)           #It will save in the file in given location
        print("\nThe Edited Image has been saved successfully !!!   :)\n")
        self.choiceMenu()
```
